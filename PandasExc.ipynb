```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
data.head()
```

```python
data['Sex'] = data['Sex'].map({'male':'M','female':'F'})
data.head()
```

```python
data['NameLen'] = data['Name'].apply(lambda x:len(x))
```

```python
data.head()
```

```python
data[('SibSp')]=data[('SibSp')].apply(np.ceil)
```

```python
data.head()
```

```python
data['NameFirst']=data['Name'].str.split(',').apply(lambda x:x[0])
```

```python
 data
```

```python
def StringPosition(str,position):
    return str[position]
```

```python
data['NameFirst']=data['Name'].str.split(',').apply(StringPosition,position=0)

data.head()
```
